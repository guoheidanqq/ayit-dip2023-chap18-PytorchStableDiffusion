# in win12

```python
git config --global --unset http.proxy
#git config --global --unset https.proxy
```

```python
git config --global http.proxy "socks5h://127.0.0.1:1080"
#git config --global https.proxy "socks5h://127.0.0.1:1080"
```

```python
git config --global https.proxy
#git config --global http.proxy
```

```python
git -c http.proxy="socks5h://127.0.0.1:1080" lfs install
#git -c https.proxy="socks5h://127.0.0.1:1080" lfs install
```

```python
# Set the 'Skip' rule and use the SOCKS proxy for the clone
$env:GIT_LFS_SKIP_SMUDGE=1
git -c http.proxy="socks5h://127.0.0.1:1080" clone https://huggingface.co/datasets/fusing/fill50k
```

```python
cd fill50k

# List all LFS files to see their status
git lfs ls-files
```

```python
# Example: Pulling just the first training zip
git -c http.proxy="socks5h://127.0.0.1:1080" lfs pull --include="*.zip"
```

```python
# Check the size of the file you just pulled
ls -lh *.zip
```
